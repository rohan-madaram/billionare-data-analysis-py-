```python
top_10 = b_data.head(10)
```

```python
industry_counts = top_10['industry'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Industry Distribution')
plt.show()
```

```python
plt.figure(figsize=(23, 6))
sns.barplot(x='full_name', y='wealth', data=top_10)


plt.xlabel('Full Name')
plt.ylabel('Wealth')
plt.title('Bar Chart of Top 10 Billionaies')
plt.show()
```

```python
industry_counts = b_data['industry'].value_counts()
top_industries = industry_counts.reset_index()
top_industries.columns = ['Industry','Counts']
print(top_industries)
```

```python
plt.figure(figsize=(8, 6))
barplot =sns.barplot(x='Counts', y='Industry', data=top_industries)

for p in barplot.patches:
    width = p.get_width()
    plt.text(width + 3,
             p.get_y() + p.get_height() / 2,
             f'{int(width)}',
             va='center')
plt.xlabel('Counts')
plt.ylabel('Industry')
plt.title('Bar Chart of Industies')
plt.show()
```

```python
countries = b_data['country_of_residence'].value_counts()
countries_count = countries.reset_index()
countries_count.columns = ['Country','Counts']
top_10_countries_of_billionaires = countries_count.head(10)
print(top_10_countries_of_billionaires)
```

```python
plt.figure(figsize=(8, 6))
barplot_two =sns.barplot(x='Counts', y='Country', data=top_10_countries_of_billionaires)

for p in barplot_two.patches:
    width = p.get_width()
    plt.text(width + 3,
             p.get_y() + p.get_height() / 2,
             f'{int(width)}',
             va='center')
plt.xlabel('Counts')
plt.ylabel('Country of Residence')
plt.title('Bar Chart of Coutries of Residence of Billionaires')
plt.show()
```
